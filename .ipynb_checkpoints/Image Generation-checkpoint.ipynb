# Tweets Timeline

```python
df["datetime"] = ["{} {}".format(vals[0], vals[1]) for vals in df[["date","time"]].values]
df.datetime = pd.to_datetime(df.datetime)
df["datetime_hour"] = [datetime.strftime(df.datetime.iloc[t], "%d-%m-%y %H") for t in range(len(df.datetime.values))]
```

```python
df.sort_values(by="datetime", inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.set(font_scale=1.5) 
sns.set_style("whitegrid")

g = sns.barplot(x="index", y="date", data=df['date'].value_counts().reset_index())
ax.set_title('Tweets per day', fontsize = 17)
g.set_xticklabels([datetime.strftime(i, "%d-%m-%y") for i in df['date'].value_counts().reset_index()["index"]])
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
filt_day = df[(df['date'] > "2020-06-13") & (df["date"] <= "2020-06-15")]
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.set(font_scale=1.5) 
sns.set_style("whitegrid")
ax.set_title('Hourly tweets since June 14 to June 15', fontsize = 17)
g=sns.lineplot(x="index", y="datetime_hour", data=filt_day['datetime_hour'].value_counts().reset_index())

g.set_xticklabels(filt_day['datetime_hour'].value_counts().reset_index()["index"].apply(lambda x: x.split(" ")[0].split("-")[0]).sort_values())
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
font_path = 'dk_lemon_yellow_sun/DK Lemon Yellow Sun.otf'
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open(path.join(d, "hija3.png")))

stopwords = set(STOPWORDS)
stopwords.add("said")
```

```python
# create coloring from image
image_colors = ImageColorGenerator(alice_mask)
```

```python

wc = WordCloud(max_words=100, mask=alice_mask, font_path=font_path,
               stopwords=stopwords, background_color="white", max_font_size=120,
               random_state=42, width=alice_mask.shape[1], collocations=False, contour_width=3, contour_color='white',
               height=alice_mask.shape[0])


# generate word cloud
wc.generate(hashtags_text)
plt.figure(figsize=(20,18))
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()

# store to file
# wc.to_file(path.join(d, "hija_output.png"))
```
